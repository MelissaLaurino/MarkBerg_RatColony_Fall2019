<b> Getting Started: Import libraries and data

```python
#Import necessary libraries:
import pandas as pd #Data import
import numpy as np #Math functions
import matplotlib.pyplot as plt #Graphing library

#Import data using Pandas:
rats = pd.read_excel('MarkBergRatColony.xlsx')
print(rats)
```

<b>How many times was each rat weighed?

```python
Rat_ID = rats.groupby('RATID')
Rat_ID_totals = Rat_ID.size()
#print (Rat_ID_totals.head())
m = Rat_ID_totals.plot(kind='bar')
m.set_xlabel("Rat ID")
m.set_ylabel("Number of times weighed")
m.set_title("Mark Berg - Animal Lab")
m
print ('Each rat weighed:')
print (Rat_ID_totals.head())
print ()
print ('Each rat was weighed 110 times during August 8, 2018 - November 26, 2018.')
```

<b>What was the range of each rat's weight?

```python
plt.figure(figsize=(20,8))
plt.plot(rats.RATID, rats.Weight_g,'.')
plt.xlabel('Rat ID', fontsize=16)
plt.ylabel('Weight (g)', fontsize=16)
plt.legend()
plt.title('Mark Berg Rat Colony', fontsize=16)
#plt.savefig('C://Users//Dhairya//Desktop//MeasureDepthAmericanStar2016.png')
plt.show()
```
